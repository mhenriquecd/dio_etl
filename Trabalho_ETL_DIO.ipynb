# Extração :

```python
import pandas as pd

nomes = pd.read_csv("nomes.csv")

display(nomes)
```

# Transformação:

```python
'''
Aqui eu coloco os dados da coluna count em ordem decrescente
de uma forma que independe do gênero da pessoa e coloco filtro para os
10 nomes mais registrados

'''
ordemdecrescente = nomes.sort_values('count', ascending=False).head(10)

# Agora crio uma tabela em ordem alfabética

alpha = nomes.sort_values('name')
```

# Carregamento:

Agora eu crio novos arquivos CSV com a tabela transformada

```python
# Os dez nomes mais usados independente do Sexo

ordemdecrescente.to_csv('emordem.csv')
```

```python
# Em Ordem alfabética

alpha.to_csv('alpha.csv')
```
